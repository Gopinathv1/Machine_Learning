```python
# A company claims that the average time taken to resolve customer tickets is 30 minutes. 
# A random sample of 40 tickets shows a mean resolution time of 33 minutes. 
# At the 5% level of significance, test whether the average resolution time has increased.

import numpy as np
from scipy import stats

np.random.seed(42)

sample_mean = 33
n = 40
population_mean = 30


sample_data = np.random.normal(loc=sample_mean, scale=10, size=n)

t_test, p_value = stats.ttest_1samp(sample_data, population_mean)

p_value_one_tailed = p_value / 2

print("T-statistic:", t_stat)
print("P-value (one-tailed):", p_value_one_tailed)

alpha = 0.05

if p_value_one_tailed < alpha and t_stat > 0:
    print("Reject H0: New course improves performance")
else:
    print("Fail to reject H0")
```

```python
# A company claims that 60% of its users are satisfied with its product.
# In a survey of 200 users, 134 users report satisfaction.
# At the 5% level of significance, test whether customer satisfaction is higher than claimed.
from statsmodels.stats.proportion import proportions_ztest


z_stat, p_value = proportions_ztest(count=134, nobs=200, value=0.60)

# Convert to one-tailed p-value
p_value_one_tailed = p_value / 2

print("Z-statistic:", z_stat)
print("P-value (one-tailed):", p_value_one_tailed)

# Decision
alpha = 0.05
if p_value_one_tailed < alpha and z_stat > 0:
    print("Reject H0: Satisfaction is higher than 60%")
else:
    print("Fail to reject H0")
```

```python
# Historically, 40% of applicants pass an entrance test.
# After a new preparation program, 90 out of 180 applicants pass.
# Test at the 5% level of significance whether the pass rate has improved.

from statsmodels.stats.proportion import proportions_ztest

z_stat, p_value = proportions_ztest(count=90, nobs=180, value=0.40)

# Convert to one-tailed p-value
p_value_one_tailed = p_value / 2

print("Z-statistic:", z_stat)
print("P-value (one-tailed):", p_value_one_tailed)

# Decision
alpha = 0.05
if p_value_one_tailed < alpha and z_stat > 0:
    print("Reject H0: Satisfaction is higher than 40%")
else:
    print("Fail to reject H0")


```

```python
import numpy as np
from scipy import stats

# Sample sizes
n1 = 30
n2 = 28

# Simulate sample data (replace with real scores in practice)
np.random.seed(42)

# Assume some means for demonstration
method_A = np.random.normal(loc=75, scale=10, size=n1)
method_B = np.random.normal(loc=70, scale=10, size=n2)

# Perform two-sample t-test (independent samples)
t_stat, p_value = stats.ttest_ind(method_A, method_B, equal_var=False)

print("T-statistic:", t_stat)
print("P-value (two-tailed):", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject H0: There is a significant difference between methods")
else:
    print("Fail to reject H0")
```

```python
import numpy as np
from scipy.stats import chi2_contingency

# Example contingency table
# Rows = Gender, Columns = Product Preference
table = np.array([
    [50, 30, 20],   # Male
    [45, 35, 25]    # Female
])

chi2, p_value, dof, expected = chi2_contingency(table)

print("Chi-square:", chi2)
print("P-value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject H0: Variables are dependent")
else:
    print("Fail to reject H0: Variables are independent")
```

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (example)
observed = np.array([18, 22, 20, 15, 25, 20])

# Expected frequencies (fair die → equal probability)
expected = np.array([20, 20, 20, 20, 20, 20])

# Chi-square test
chi2, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square:", chi2)
print("P-value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject H0: Die is not fair")
else:
    print("Fail to reject H0: Die is fair")
```
